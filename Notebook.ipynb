<h1 align="center">Lane Detection</h1>
<p align="center">Bài toán nhận dạng làn đường, vị trí của xe trong làn đường. Để tối ưu trên Jetson Nano, ta sử dụng các phương pháp xử lí ảnh. Do chỉ có một mình xe chạy và đường đơn giản nên không cần sử dụng Neural Network.</p>

ref: https://automaticaddison.com/the-ultimate-guide-to-real-time-lane-detection-using-opencv/

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('./data/lane.jpeg')
```

```python
# height, width, channel
h, w, c = img.shape
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

Ta cắt ảnh chỉ lấy phần có làn đường. Gọi là vùng quan tâm(ROI - Region Of Interest)

```python
roi_img = img.copy()
roi_img[:h//2, :] = np.zeros_like(roi_img[:h//2, :])

plt.imshow(cv2.cvtColor(roi_img, cv2.COLOR_BGR2RGB))
plt.show()
```

Để nhận dạng được làn đường đang đi ta cần nhận dạng hai vạch kẻ ở hai bên của làn đường.

```python
gray_img = cv2.cvtColor(roi_img, cv2.COLOR_BGR2GRAY)
rgb_img  = cv2.cvtColor(roi_img, cv2.COLOR_BGR2RGB)
hsv_img  = cv2.cvtColor(roi_img, cv2.COLOR_BGR2HSV)
hls_img  = cv2.cvtColor(roi_img, cv2.COLOR_BGR2HLS)
```

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle('Image in different color space')
axs[0, 0].imshow(rgb_img)
axs[0, 0].set_title('RGB')
axs[0, 1].imshow(gray_img, cmap='gray', vmin=0, vmax=255)
axs[0, 1].set_title('GRAY')
axs[1, 0].imshow(hsv_img)
axs[1, 0].set_title('HSV')
axs[1, 1].imshow(hls_img)
axs[1, 1].set_title('HLS')

plt.show()
```
